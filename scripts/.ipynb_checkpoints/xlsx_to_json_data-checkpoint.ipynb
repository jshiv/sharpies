```python
import pandas as pd

import json

from pygeocoder import Geocoder
```

```python
def getLocation(loc):

    '''
    Given a string in the format of an address or geographic coordinates (latitude, longitude), returns a tuple of size 2
    containing:

    * a string giving the formatted address
    * a tuple of size 2 giving the geographic coordinates (latitude, longitude)

    For more information, see: http://www.anmolgarg.com/mapping-community-engagement/

    Parameters
    ----------
    loc : string in the format of an address, or 2-entry dictionnary which values are the latitude and longitude
        Location in the world to translate

    Returns
    -------
    address, (lat, lon) : tuple
        python tuple containing the address and the coordinates. 

    '''

    r = Geocoder.geocode(loc)
    address = r.raw[0]['formatted_address']
    location = r.raw[0]['geometry']['location']
    return address, (location['lat'],location['lng'])
```

```python
df = pd.read_excel("./data/SF nightclubs and bars - final (Ben Drobiz's conflicted copy 2014-11-23).xlsx",sheetname = 'Sheet3', header = 2)
```

```python
for col in df.columns:
    print col.lower()
    df.rename(columns = {col:col.lower()}, inplace = True )
    col = col.lower()
    if 'image' in col.lower():
        df.rename(columns = {col:'image'}, inplace = True )
        
    if 'open' in col.lower():
        df.rename(columns = {col:'hours'}, inplace = True )
        
    if 'long' in col.lower():
        df.rename(columns = {col:'long'}, inplace = True )

    if 'lat' in col.lower():
        df.rename(columns = {col:'lat'}, inplace = True )

```

```python
df = df[df.address.isnull() == False]
df = df.reset_index(drop = True)
df.drop(labels = ['picture'], axis = 1, inplace = True)
df['lat'] = ''
df['long'] = ''
df.head()
```

```python
for idx in df.index:
    try:
        address, (lat, lng) = getLocation(df['address'][idx])
        df['address'][idx] = address
        df['lat'][idx] = str(lat)
        df['long'][idx] = str(lng)
        df['type'][idx] = df['type'][idx].title()
    except:
        pass
```

```python
df
```

```python
places = []
for idx in df.index:
    places.append(df.ix[idx].to_dict())
```

```python
places
```

```python
import re
def remove_non_ascii_1(text, rep = ''):

    return ''.join([i if ord(i) < 128 else rep for i in text])
def remove_non_ascii_2(text, rep = ''):

    return re.sub(r'[^\x00-\x7F]+',rep, text)

for place in places:
    for k,v in place.iteritems():
        if k == 'hours':
            place[k] = remove_non_ascii_2(remove_non_ascii_1(v, rep = '-'), rep = '-')
        else:
            place[k] = remove_non_ascii_2(remove_non_ascii_1(v, rep = ''), rep = '')

```

```python
with open('./data/places.json', 'w') as f:
    json.dump(places, f)
```
